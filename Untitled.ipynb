**entropy function**

```python
def entropy(p):  #here p is the fraction of positive results or fraction of cat
    if p == 0 or p == 1: #if the p is 0 then it means the splitted node has either all cats or all dogs. so the impurity level is 0 or 1 depending upon whether its cat(positive result) or dog(negative result).
        return 0
    else:
        return -p * np.log2(p) - (1-p) * np.log2(1-p)
print(f'The entropy of the parent root node of the decision tree is: {entropy(0.5)}')  #as there are 5 cats and 5 dogs in the training sets        
```

**Splittng function based on the feature, which produces not the result directly but the index of the result of array y**

```python
def feature_based_split(X,feature):
    left_node = []
    right_node = []
    for i,x in enumerate(X):
        if x[feature] == 1:  #if the current traning example has this feature then we push it to left node
            left_node.append(i)
        else:
            right_node.append(i)  #otherwise we push it to right_node
    return left_node, right_node
print(f'The left node and right node after splitting the parent node using feature earshape is: {feature_based_split(X_train,0)} respectively')      
```

**Weighted Entropy**

```python
def weighted_entropy(X,y,left_node,right_node):
    pleft = sum(y[left_node]) / len(left_node)  #fraction of positive result in leftnode
    pright = sum(y[right_node]) / len(right_node)  #fraction of positive result in rightnode
    wleft = len(left_node) / len(X)  #fraction of datas in leftnode out of total datasets
    wright = len(right_node) / len(X) #fraction of datas in rightnode out of total datasets
    return (wleft * pleft) + (wright * pright)
    
```
